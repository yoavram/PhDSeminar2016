```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('white')
sns.set_context('talk', font_scale=1.5)
```

```python
import scipy.special as sp
```

```python
n = 7
r = 0.01
d = np.linspace(r, 0.25, 1000)
p = lambda n, r, d: 1 - sp.erf(np.sqrt(n) * r / d)
```

```python
plt.plot(d, [p(3, r, _d) for _d in d], label='n=3')
plt.plot(d, [p(7, r, _d) for _d in d], label='n=7')
plt.plot(d, [p(15, r, _d) for _d in d], label='n=15')
plt.xlabel('Distance to optimum - d')
plt.ylabel('Beneficial mutation probability - p')
plt.legend(loc='lower right')
sns.despine()
```

```python
r = np.linspace(0.01, 1)
plt.plot(d, [p(3, _r, 1) for _r in d], label='n=3')
plt.plot(d, [p(7, _r, 1) for _r in d], label='n=7')
plt.plot(d, [p(15, _r, 1) for _r in d], label='n=15')
plt.xlabel('Mutation size - r')
plt.ylabel('Beneficial mutation probability - p')
plt.legend()
sns.despine()
```
